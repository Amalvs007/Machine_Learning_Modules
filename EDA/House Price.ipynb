### R-squared & Adjusted R-squared

```python
X_train.head()
```

```python
r2_score=[]
adjusted_r2=[]
feature_names=[]

for i in range(1,20):
    feature_names.append(X_train.columns[:i].to_list())
    lr1=LinearRegression()
    lr1.fit(X_train[X_train.columns[:i]],y_train)
    R2=lr1.score(X_test[X_test.columns[:i]],y_test)
    r2_score.append(R2)

    n=len(X_train)
    p=i
    adjR2=1-((1-R2) * (n-1) / (n-p-1))
    adjusted_r2.append(adjR2)
```

```python
df2=pd.DataFrame(np.column_stack((r2_score,adjusted_r2)),columns=['R2','Adjusted R2'])
df2['Feature Names']=feature_names
```
