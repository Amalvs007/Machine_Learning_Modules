```python
# Legend, Grid
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather Info')

plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')
#plt.legend()
plt.legend(loc="best",shadow=True,fontsize='small')
plt.grid()
plt.show()
```

```python
#Bar chart

company=['Google','Amazon','Microsoft','Facebook']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
plt.ylabel('Revenue')
plt.xlabel('Companies')
plt.title('Revenue Model for Top Companies')
plt.bar(company,revenue,label='Revenue')
plt.bar(company,profit,label='Profit')
plt.legend()
plt.show()
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company) # For changing name of the values
plt.ylabel('Revenue')
plt.xlabel('Companies')
plt.title('Revenue Model for Top Companies')
plt.bar(xpos-0.2,revenue,width=0.4,label='Revenue')
plt.bar(xpos+0.2,profit,width=0.4,label='Profit')
plt.legend()
plt.show()
```

```python
plt.barh(company,revenue)
```

```python
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]

plt.xlabel('Sugar Level')
plt.ylabel('Number of Patients')
plt.title('Charts')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.8,label=['Men','Women'],color=['red','black'])
plt.legend()
plt.show()
```

```python
# Box Plot

np.random.seed(10)
data=np.random.normal(100,20,200)
data
```

```python
plt.hist(data,rwidth=0.8)
```

```python
plt.figure(figsize=(10,5))
plt.boxplot(data)
plt.show()
```

```python
np.random.seed(10)
data1=np.random.normal(100,10,200)
data2=np.random.normal(90,20,200)
data3=np.random.normal(80,30,200)
data4=np.random.normal(70,40,200)

data=[data1,data2,data3,data4]
plt.figure(figsize=(10,5))

plt.boxplot(data,vert=False,patch_artist=True)
plt.show()
```

```python
# Pie Chart

exp_values=[1400,600,300,410,250]
exp_labels=['Home Rent','Food','Phone/Internet Bill','car','Other Utiliteis']
```

```python
plt.figure(figsize=(10,7))
plt.axis('equal')
plt.pie(exp_values,labels=exp_labels,radius=1.2,autopct="%0.3f%%",shadow=True,explode=[0.1,0,0.1,0,0.4],startangle=45)
plt.legend()
# plt.show()
plt.savefig('piepic.jpg') # The extension in the name will decide which the file tyoe is and it is according to our preferenece
```
