```python
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.datasets import load_boston
import warnings
warnings.filterwarnings('ignore')
```

```python
# Boston is a dictionary.
```

```python
boston = load_boston()
data = pd.DataFrame(boston.data,columns=boston.feature_names)
data['Price'] = boston.target
data.head()
```

```python
X = data.drop(['Price'],axis=1)
y = data['Price']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
dt_Regressor = DecisionTreeRegressor()
dt_Regressor.fit(X_train,y_train)

R_squared_Train=dt_Regressor.score(X_train,y_train)
print('R_squared_Train =',R_squared_Train)

R_squared_Test = dt_Regressor.score(X_test,y_test)
print('R_squared_Test =',R_squared_Test)

ypred = dt_Regressor.predict(X_test)

mse = mean_squared_error(y_test,ypred)
print('MSE =' ,mse)

RMSE = mse**(1/2.0)
print('RMSE =',RMSE)
```

```python
from sklearn import tree
plt.Figure(figsize=(20,15))
tree.plot_tree(dt_Regressor,filled=True,class_names=True)
plt.show()
```

```python
dt_Regressor = DecisionTreeRegressor(max_depth=7)
dt_Regressor.fit(X_train,y_train)

R_squared_Train=dt_Regressor.score(X_train,y_train)
print('R_squared_Train =',R_squared_Train)

R_squared_Test = dt_Regressor.score(X_test,y_test)
print('R_squared_Test =',R_squared_Test)

ypred = dt_Regressor.predict(X_test)

mse = mean_squared_error(y_test,ypred)
print('MSE =' ,mse)

RMSE = mse**(1/2.0)
print('RMSE =',RMSE)
```
